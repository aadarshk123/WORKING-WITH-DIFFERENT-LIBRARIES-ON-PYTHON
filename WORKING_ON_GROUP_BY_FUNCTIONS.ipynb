```python
import pandas as pd
data=pd.read_csv("Quest8.csv")
df=pd.DataFrame(data)
df
```

```python
#PART A
#FOR DISPLAYING GROPED VALUE WE NEED TO APPLY apply() func

df.groupby('Name').apply(display)
```

```python
#PART A
df.groupby('Name').sum()
```

```python
#PART A (2.0)
df.sort_values('Name').groupby('Name').sum()
```

```python
#PART B
#MINIMUM,MAXIMUM,TOTAL MONTHLY INCOME OF EACH FAMILY

df.groupby('Name')['MonthlyIncome(Rs.)'].agg(func=['min','max','sum'])
```

```python
#PART D
#help(df.agg)

df[df['MonthlyIncome(Rs.)'] >= 60000.00]
```

```python
#PART D (2.0)
df[df['MonthlyIncome(Rs.)'] >= 60000.00].sort_values('Name')
```

```python
#PART C
temp1=df.groupby('Name').apply(lambda x: x[x['MonthlyIncome(Rs.)']==x['MonthlyIncome(Rs.)'].max()])
temp1.reset_index(inplace=True,drop=True)
temp1
```

```python
#PART E

temp1=df[(df['Name']=="Shah") & (df['Gender']=="Female")]
print("The average monthly income of the female members in the Shah family : ")
temp1['MonthlyIncome(Rs.)'].mean()
```
