```python
# Create a dataframe having at least 3 columns and 50 rows to store numeric data generated using a random
# function. Replace 10% of the values by null values whose index positions are generated using random function.
# Do the following:
# a. Identify and count missing values in a dataframe.
# b. Drop the column having more than 5 null values.
# c. Identify the row label having maximum of the sum of all values in a row and drop that row.
# d. Sort the dataframe on the basis of the first column.
# e. Remove all duplicates from the first column.
# f. Find the correlation between first and second column and covariance between second and third
# column.
# g. Detect the outliers and remove the rows having outliers.
# h. Discretize second column and create 5 bins


import numpy as np
import pandas as pd
from pandas import Series,DataFrame
#from itertools import zip

rand = np.random.randn(50,3)
df=DataFrame(rand)
temp=DataFrame(rand)                ##For storing original data values
print("The DataFrame is : ")
df
```

```python
per=(10/100)*df.size
#print("\n",per)

for i in range(int(per)):
    row=np.random.randint(0,50)
    col=np.random.randint(0,3)
    df.loc[row,col]=np.nan
    while(df.loc[row,col] is np.nan):
        row=np.random.randint(0,50)
        col=np.random.randint(0,3)
    df.loc[row,col]=np.nan
        
print("\nThe new DataFrame is : ")
df
```

```python
#PART A
#print("\nTotal number of N.A. values in dataframe : ",df.isnull().sum())                          --> print N.A. value columnwise 
print("\nTotal number of N.A. values in dataframe : ",df.isnull().sum().sum())
```

```python
#PART B
df.dropna(thresh=45,axis=1,inplace=True)
print("\nThe DataFrame after dropping column with 5 N.A. value is : ")
df
```

```python
#PART C
print("\nSum of each row of dataframe : ")
print(temp.sum(axis=1))

ind=np.argmax(temp.sum(axis=1))
print("\nIndex with maximum sum : ",ind)
temp.drop(ind)
```

```python
#PART D
print("\nOn sorting Dataframe values according to first column : ")
temp.sort_values(by=[temp.columns[0]],inplace=True)
temp
```

```python
print("\nOn sorting Dataframe values according to first and second column : ")
temp.sort_values(by=[temp.columns[0],temp.columns[1]],inplace=True)
temp
```

```python
#PART E
print("\nOn remove all duplicates from the first column : ")
temp.drop_duplicates(subset=df.columns[0],keep="first",inplace=True)
temp

# help(df.drop_duplicates)
```

```python
#PART F
cor=temp[temp.columns[0]].corr(temp[temp.columns[1]])
print("\nCorrelation between first and second column : ",cor)

cov=temp[temp.columns[1]].cov(temp[temp.columns[2]])
print("\nCovariance between second and third column : ",cov)
```

```python
#PART G
Q1=temp[0].quantile(0.25)
Q3=temp[0].quantile(0.75)

IQR=Q3-Q1

print(Q3+(1.5*IQR))
print(Q1-(1.5*IQR))

nrow={0:[10000,-10000],1:[2,2],2:[2,2]}
tempdf=pd.DataFrame(nrow)

newdf=pd.concat([temp,tempdf],ignore_index=True)
newdf
```

```python
upper=np.where(newdf[0] >= Q3+(1.5*IQR))
lower=np.where(newdf[0] <= Q1-(1.5*IQR))

newdf.drop(upper[0],inplace=True)
newdf.drop(lower[0],inplace=True)

newdf
```

```python
#PART H
Cut=pd.cut(temp[temp.columns[1]],bins=5)
print("\nUsing cut : \n")
print(Cut)
print("\nCount : \n")
Cut.value_counts()
```

```python
Qcut=pd.qcut(temp[temp.columns[1]],q=5)
print("\nUsing qcut : \n")
print(Qcut)
print("\nCount : \n")
print(Qcut.value_counts())
```
