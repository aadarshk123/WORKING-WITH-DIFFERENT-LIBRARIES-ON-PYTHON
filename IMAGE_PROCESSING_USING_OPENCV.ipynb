```python
#RESIZING AN IMAGE
img1=cv2.imread("ASIA_CUP.jpg")
cv2.imshow("ORIGINAL_IMAGE",img1)
cv2.waitKey(0)
cv2.destroyAllWindows()

resize_img=cv2.resize(img1,(300,300))
cv2.imshow("RESIZED_IMAGE",resize_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#CROPPING AN IMAGE

'''x1,y1,x2,y2 are the cordinates of ROI(Region of interest)'''

x1, y1 = 400, 400 
x2, y2 = 800, 800
cropped_image = img1[y1:y2, x1:x2]

cv2.imwrite('cropped_image_AsiaCup.jpg', cropped_image)

cv2.imshow('Cropped Image', cropped_image)
cv2.waitKey(0)  
cv2.destroyAllWindows()
```

```python
#ROTATING AN IMAGE

"""image.shape will return 3 features 
1.Height (number of rows)
2.Width (number of columns)
3.Number of channels"""

(h,w) = img1.shape[:2]
center = (h/2,w/2)

angle = 45
scale = 1.0
M = cv2.getRotationMatrix2D(center, angle, scale)

rotated_image = cv2.warpAffine(img1, M, (w, h))

cv2.imwrite('rotated_image_AsiaCup.jpg', rotated_image)

cv2.imshow('Rotated Image', rotated_image)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
#FLIPPING AN IMAGE

# Flip horizontally
flipped_horizontally = cv2.flip(img1, 1)

# Flip vertically
flipped_vertically = cv2.flip(img1, 0)

# Flip both horizontally and vertically
flipped_both = cv2.flip(img1, -1)


cv2.imwrite('ASIA_CUP_flipped_horizontally.jpg', flipped_horizontally)
cv2.imwrite('ASIA_CUP_flipped_vertically.jpg', flipped_vertically)
cv2.imwrite('ASIA_CUP_flipped_both.jpg', flipped_both)

cv2.imshow('ASIA_CUP_Flipped Horizontally', flipped_horizontally)
cv2.imshow('ASIA_CUP_Flipped Vertically', flipped_vertically)
cv2.imshow('ASIA_CUP_Flipped Both', flipped_both)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#TRANSLATING THE IMAGE

'''The matrix [[1, 0, tx], [0, 1, ty]] translates the image tx(+ve value) pixels right and ty(+ve value) pixels down.
   The matrix [[1, 0, tx], [0, 1, ty]] translates the image tx(-ve value) pixels left and ty(-ve value) pixels Upward.'''

tx, ty = -250,-100  
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])

(height, width) = img1.shape[:2]
translated_image = cv2.warpAffine(img1, translation_matrix, (width, height))

cv2.imwrite('ASIA_CUP_translated_image.jpg', translated_image)

cv2.imshow('ASIA_CUP_Translated Image', translated_image)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
#ADJUSTING BRIGHTNESS AND CONTRAST
#METHOD-1

def adjustBrightnessAndContrast(image, brightness=0, contrast=0):
    img = image.astype(np.float32)
    img = img + brightness
    img = img * (1 + contrast / 100.0)
    img = np.clip(img, 0, 255)
    img = img.astype(np.uint8)
    return img

image = cv2.imread('ASIA_CUP.jpg')

new_img = adjustBrightnessAndContrast(image, brightness=20, contrast=40)

cv2.imshow('Original Image', image)
cv2.imshow('Adjusted Image', new_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
    
cv2.imwrite('adjusted_ASIA_CUP.jpg', new_img)
```

```python
#METHOD-2

def adjust_brightness_contrast(image, alpha=0, beta=0):
    adjusted_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
    return adjusted_image

image_path = 'WhatsApp Image 2024-07-16 at 15.57.41_ca01b74a.jpg'
image = cv2.imread(image_path)

alpha = 1.0  
beta = 20    

adjusted_image = adjust_brightness_contrast(image, alpha, beta)

cv2.imwrite('WhatsApp Image 2024-07-16 at 15.57.41_ca01b74a_adjusted_image.jpg', adjusted_image)

cv2.imshow('Original Image', image)
cv2.imshow('Adjusted Image', adjusted_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#HISTOGRAM EQUALIZATION

def equalize_histogram(image):
    if len(image.shape) == 3:
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray_image = image

    equalized_image = cv2.equalizeHist(gray_image)
    return equalized_image

image_path = 'ASIA_CUP.jpg'
image = cv2.imread(image_path)

equalized_image = equalize_histogram(image)

cv2.imwrite('equalized_image_ASIA_CUP.jpg', equalized_image)

cv2.imshow('Original Image', image)
cv2.imshow('Equalized Image', equalized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#DENOISING TECHNIQUES
#1.GAUSSIAN SMOOTHING

image = cv2.imread('ASIA_CUP.jpg')

gaussian_blur = cv2.GaussianBlur(image, (5, 5), 0)

cv2.imshow('Original Image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imshow('ASIA_CUP_Gaussian Smoothing', gaussian_blur)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite('ASIA_CUP_gaussian_smoothing.jpg', gaussian_blur)
```

```python
#2.MEDIAN SMOOTHING

image = cv2.imread('ASIA_CUP.jpg')

median_blur = cv2.medianBlur(image, 5)

cv2.imshow('Original Image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imshow('ASIA_CUP_Median Smoothing', median_blur)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite('ASIA_CUP_median_smoothing.jpg', median_blur)
```

```python
#3.Averaging
image = cv2.imread('ASIA_CUP.jpg')

average_filtered = cv2.blur(image, (5, 5))

cv2.imshow('Original _image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imshow('Averaging', average_filtered)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite('ASIA_CUP_average_filtered.jpg', average_filtered)
```

```python
#4.Non-Local Means
image = cv2.imread('ASIA_CUP.jpg')

gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

nlm_filtered = cv2.fastNlMeansDenoising(gray_image, None, 30, 7, 21)

cv2.imshow('Original _image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imshow('Non-Local Means Denoising', nlm_filtered)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Save the result
cv2.imwrite('ASIA_CUP_nlm_filtered.jpg', nlm_filtered)
```

```python
#5.BM3D (Block-Matching and 3D filtering)
import bm3d

image = cv2.imread('ASIA_CUP.jpg', -1)

bm3d_filtered = bm3d.bm3d(image, sigma_psd=5, stage_arg=bm3d.BM3DStages.ALL_STAGES)

cv2.imshow('Original _image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imshow('BM3D Denoising', bm3d_filtered)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite('ASIA_CUP_bm3d_filtered.jpg', bm3d_filtered)
```

```python
#SHARPENING

image = cv2.imread('WhatsApp Image 2024-07-16 at 15.57.41_ca01b74a.jpg')

sharpening_kernel = np.array([[-1, -1, -1], 
                              [-1,  9, -1],
                              [-1, -1, -1]])

sharpened_image = cv2.filter2D(image, -1, sharpening_kernel)

cv2.imshow('Original Image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imshow('WhatsApp Image 2024-07-16 at 15.57.41_ca01b74a.jpg_Sharpening', sharpened_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite('ASIA_CUP_sharpened_image.jpg', sharpened_image)
```
