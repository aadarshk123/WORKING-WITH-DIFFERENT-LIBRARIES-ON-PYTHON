```python
# Taking Iris data, plot the following with proper legend and axis labels: (Download IRIS data from:
# https://archive.ics.uci.edu/ml/datasets/iris or import it from sklearn.datasets)
# a. Plot bar chart to show the frequency of each class label in the data.
# b. Draw a scatter plot for Petal width vs sepal width.
# c. Plot density distribution for feature petal length.
# d. Use a pair plot to show pairwise bivariate distribution in the Iris Dataset.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.datasets import load_iris
df=load_iris()
df 
```

```python
dir(df)
#print(df.target)
```

```python
iris=pd.DataFrame(df.data,columns=df.feature_names)
iris
```

```python
iris['species']=df.target
print("\nDataSet after adding 'species' column : \n")
iris
```

```python
cor=iris['sepal length (cm)'].corr(iris['sepal width (cm)'])
cor
```

```python
iris['species']=iris['species'].replace([0],'Setosa')
iris['species']=iris['species'].replace([1],'Versicolour')
iris['species']=iris['species'].replace([2],'Virginica')
print("\nDataSet after adding 'species' column and replcing the values : \n")
iris
```

```python
#PART A

freq=iris['species'].value_counts()
freq.plot(kind="bar",color={"red","green","blue"})
freq
```

```python
#PART A using seaborn library

sns.countplot(x="species",data=iris)
```

```python
#PART B
plt1=iris[iris["species"]=="Setosa"].plot(kind="scatter",x="petal width (cm)",y="sepal width (cm)",color="blue",marker="o",label="Setosa")
plt.grid()
plt.title("Scatter Plot for species")

plt2=iris[iris["species"]=="Versicolour"].plot(kind="scatter",x="petal width (cm)",y="sepal width (cm)",color="red",marker="*",label="Versicolour",ax=plt1)
plt.grid()
plt.title("Scatter Plot for spaecies")

plt3=iris[iris["species"]=="Virginica"].plot(kind="scatter",x="petal width (cm)",y="sepal width (cm)",color="green",marker="+",label="Virginica",ax=plt1)
plt.grid()
plt.title("Scatter Plot for species")
```

```python
#PART B using seaborn library

#import seaborn as sns
sns.scatterplot(x="petal width (cm)",y="sepal width (cm)",data=iris,hue="species")
```

```python
#PART C

iris["petal length (cm)"].plot(kind="density")
plt.xlabel("petal length")
plt.grid()
plt.show()
```

```python
#PART C USING SEABORN library

sns.kdeplot(iris['petal length (cm)'])
plt.grid()
```

```python
#PART D

sns.pairplot(data=iris,hue="species")
```
